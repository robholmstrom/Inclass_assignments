Write a function to compute entropy given an input of a sequence of probabilities.

```python
#answer below:

from math import log2

def ent(x):
    result = 0
    for i in x:
        result = result + -1*log2(i)*i
    return result
```

Aggregate the tennis decision table for each value of each column. Start with Outlook below. Compute the weighted mean of the entropy for outlook (the weighted mean of the yes decision and the no decision).

```python
def ent_weight(df, feature, category, outcome):
    cat_count = df[df[feature] == category].groupby([outcome])[outcome].count()
    cat_sum = np.sum(df[feature] == category)
    entropy = ent(cat_count / cat_sum)
    weight = cat_sum / df.shape[0]
    # Uncomment the following lines for debugging:
    # print('Breakdown for ' + category + ':\n' + str(cat_count))
    # print('Category weight: ' + str(cat_sum) + '/' + str(df.shape[0]) + '\n')
    return entropy, weight
```

```python
def sum_ent(df, feature, outcome):
    categories = df[feature].unique()
    total = 0.0
    for category in categories:
        entropy, weight = ent_weight(df, feature, category, outcome)
        total += entropy * weight
    return total
```

```python
sum_ent(tennis, 'Outlook', 'Decision')
```

Compute the weighted mean of the entropy for temperature, humidity and wind as well and decide based on these values which should be the first variable chosen for a split.

```python
# answer below:

results = pd.Series(dtype=float)

results['Outlook'] = sum_ent(tennis, 'Outlook', 'Decision')
results['Temp.'] = sum_ent(tennis, 'Temp.', 'Decision')
results['Humidity'] = sum_ent(tennis, 'Humidity', 'Decision')
results['Wind'] = sum_ent(tennis, 'Wind', 'Decision')

results
```
