In this assignment, we will learn statistical significance in linear models. We will use the google play store dataset loaded below and analyze the regression from this dataset.

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
reviews = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/googleplaystore.csv')
```

```python
reviews.head()
```

```python
reviews.info()
```

We will predict app ratings using other features describing the app. To use these features, we must clean the data.

Start by creating dummy variables out of the type and content rating columns.

```python
# answer below:

type_ = pd.get_dummies(reviews['Type'])
type_.drop('Free', axis=1, inplace=True)

content_rating = pd.get_dummies(reviews['Content Rating'])
content_rating.drop('Everyone', axis=1, inplace=True)
```

```python
reviews_clean = pd.concat([reviews.drop(['Type', 'Content Rating'], axis=1), type_, content_rating], axis=1)
reviews_clean.head()
```

Next, check for missing values and remove all rows containing missing values

```python
# answer below:
reviews_clean.info()
```

```python
reviews_clean.dropna(inplace=True)
```

```python
reviews_clean.info()
```

To simplify, we will remove the app, category, size, installs, genres, last updated, current ver, and android ver columns. 

```python
# answer below:
reviews_clean.drop(['App', 'Category', 'Size', 'Installs', 'Genres', 'Last Updated', 'Current Ver', 'Android Ver'], axis=1, inplace=True)
reviews_clean.head()
```

Next, check that all the columns are of numeric type and change the type of columns that are not numeric. If coercing to numeric causes missing values, remove those rows containing missing values from our dataset.

```python
# answer below:

reviews_clean.info()
```

```python
reviews_clean['Reviews'] = pd.to_numeric(reviews_clean['Reviews'])
```

```python
reviews_clean.drop(reviews_clean.loc[(reviews_clean['Price']  == 'Everyone')].index, inplace=True)

reviews_clean.loc[(reviews_clean['Price'] != '0'), 'Price'] = reviews_clean.loc[(reviews_clean['Price'] != 0), 'Price'].str.slice(1)
reviews_clean.loc[:, 'Price']= pd.to_numeric(reviews_clean.loc[:, 'Price'])
reviews_clean.Price.unique()
```

```python
reviews_clean.info()
```

Perform a train test split with 20% of the data in the test sample.

```python
# answer below:
X = reviews_clean.drop('Rating', axis=1)
y = reviews_clean['Rating']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

Now generate a linear model using statsmodels or sklearn and produce a p value for each coefficient in the model. Analyze the results.

```python
#answer below:
import statsmodels.api as sm

X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)
model = sm.OLS(y_train, X_train)

results = model.fit()
```
