```python
# answer below:
dm = pd.get_dummies(csection['Delivery number'])
dt = pd.get_dummies(csection['Delivery time'])
bp = pd.get_dummies(csection['Blood Pressure'])

csdm = pd.concat([dm, dt, bp, csection.drop(['Delivery number','Delivery time','Blood Pressure'],axis = 1 )], axis = 1)
X = csdm.drop('Caesarian', axis = 1)
y = csdm['Caesarian']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .2)

clist = [0.0001, 1,10, 100, 1000, 10000, 100000]

for x in clist:
  from sklearn.svm import SVC
  svm = SVC(kernel = 'linear', C=x )
  svm.fit(X_train, y_train)
  print('For C value of {} the test score is '.format(x) ,svm.score(X_test, y_test))
  print(svm.score(X_train, y_train))
```

What if we switched up the target variable? Let assume that we know whether a patient received a c section. Let's try to predict whether she has high blood pressure. 

Create the SVC model with a linear kernel with Blood Pressure as the target variable. Report both the train and the test scores.

```python
dm = pd.get_dummies(csection['Delivery number'])
dt = pd.get_dummies(csection['Delivery time'])
bp = pd.get_dummies(csection['Caesarian'])

csdm = pd.concat([dm, dt, bp, csection.drop(['Delivery number','Delivery time','Caesarian'],axis = 1 )], axis = 1)
X = csdm.drop('Blood Pressure', axis = 1)
y = csdm['Blood Pressure']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = .2)
```

```python
# answer below:
clist = [0.1, 1,10, 100, 1000, 10000, 100000]

for x in clist:
  from sklearn.svm import SVC
  svm = SVC(kernel = 'linear', C=x )
  svm.fit(X_train, y_train)
  print('For C value of {} the test score is '.format(x) ,svm.score(X_test, y_test))
  print(svm.score(X_train, y_train))
```

Show confusion matrices for the training and test sets, and a classification report for the test set. What trends do you notice?

```python
from sklearn.metrics import confusion_matrix
y_pred = svm.predict(X_train)
confusion_matrix(y_train, y_pred)
```

```python
y_pred_test = svm.predict(X_test)
confusion_matrix(y_test, y_pred_test)
```
