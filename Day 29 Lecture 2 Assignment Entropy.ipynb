Write a function to compute entropy given an input of a sequence of probabilities.

```python
#answer below:
from math import log2

def entropy(df):
  dfname = str(df)[0:str(df).index(' ')]
  entno = 0
  entyes = 0
  for x in df.iloc[0].tolist():
     if x != 0:
      entno += -x/df[df.index =='No'].sum().sum()*log2(x/df[df.index =='No'].sum().sum())
  for x in df.iloc[1].tolist():
     if x != 0:
      entyes += -x/df[df.index =='Yes'].sum().sum()*log2(x/df[df.index =='Yes'].sum().sum())
  mean = entno*(sum(df.iloc[0].tolist())/df.sum().sum()) + entyes*(sum(df.iloc[1].tolist())/df.sum().sum())
  
  print('The weighted mean for {} of entropy is {}'.format(dfname, mean))



#Tennisct = pd.concat([Outlook, Humidity, Wind, Temp], axis = 1)



```
